```python
#Q1
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
import plotly.express as px
fig = px.scatter(data_frame = titanic, 
                 x = "age", 
                 y = "fare", 
                 title = "Titanic Dataset: Age vs. Fare")
fig.show()
```

```python
#Q2
tips = px.data.tips()
```

```python
fig = px.box(data_frame = tips,
             x = "day",
             y = "total_bill",
             title = "dataset")
fig.show()
```

```python
#Q3
fig = px.histogram(data_frame = tips,
                   x = "sex",
                   y = "total_bill",
                   color = "day",
                   pattern_shape = "smoker",
                   title = "total bill:gender and smoking status")
fig.show()
```

```python
from pandas.core.ops import frame_arith_method_with_reindex
iris = px.data.iris()
fig = px.scatter_matrix(data_frame = iris,
                        dimensions = ["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color = "species",
                        title = "Iris Dataset: Scatter Matrix Plot")
fig.update_layout(height = 500)
fig.show()     
```

#Q5
distplot is used to plot a histogram and a kernel density estimate of a continuous variable. However, in Plotly Express, we can achieve a similar plot using the hist function or the kde function separately, but not in a combined manner as in Seaborn's distplot.

```python
tips = px.data.tips()
fig = px.histogram(data_frame = tips, 
                   x = "total_bill", 
                   y = "tip", 
                   color = "sex",
                   marginal = "box",
                   hover_data = tips.columns)
fig.show()
```
